```python
# Rename filtered df
```

```python
dffiltered = df[(df['Event distance/length'].isin(['50km', '50mi'])) & 
   (df['Year of event'] == 2020) & 
   (df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA')]
```

```python
dffiltered.head(10)
```

```python
dffiltered.shape
```

```python
# Remove (USA) from event name
```

```python
dffiltered['Event name'].str.split('(').str.get(0)
```

```python
dffiltered['Event name'] = dffiltered['Event name'].str.split('(').str.get(0)
```

```python
dffiltered.head(10)
```

```python
# Clean athlete age column
```

```python
dffiltered['athlete_age'] = 2020 - dffiltered['Athlete year of birth']
```

```python
# Remove 'h' from athlete performance
```

```python
dffiltered['Athlete performance'] = dffiltered['Athlete performance'].str.split(' ').str.get(0)
```

```python
dffiltered.head(5)
```

```python
# Drop columns: Athlete Club, Athlete Country, Athlete year of birth, Athlete Age Category
```

```python
dffiltered = dffiltered.drop(['Athlete club', 'Athlete country', 'Athlete year of birth', 'Athlete age category'], axis = 1)
```

```python
dffiltered.head(10)
```

```python
# Clean up null values
```

```python
dffiltered.isna().sum()
```

```python
dffiltered[dffiltered['athlete_age'].isna()==1]
```

```python
dffiltered = dffiltered.dropna()
```

```python
dffiltered.shape
```

```python
# Check for duplicates
```

```python
dffiltered[dffiltered.duplicated() == True]
```

```python
# Reset index
```

```python
dffiltered = dffiltered.reset_index(drop = True)
```

```python
# Fix data types
```

```python
dffiltered.dtypes
```

```python
dffiltered['athlete_age'] = dffiltered['athlete_age'].astype(int)
```

```python
dffiltered['Athlete average speed'] = dffiltered['Athlete average speed'].astype(float)
```

```python
dffiltered.dtypes
```

```python
dffiltered.head(10)
```

```python
# Rename columns
```

```python
dffiltered = dffiltered.rename(columns = {'Year of event':'year',
                                           'Event dates':'race_day',
                                           'Event name':'race_name',
                                          'Event distance/length':'race_length',
                                          'Event number of finishers':'race_number_of_finishers',
                                          'Athlete performance':'athlete_performance',
                                          'Athlete gender':'athlete_gender',
                                          'Athlete average speed':'athlete_average_speed',
                                          'Athlete ID': 'athlete_id'
                                         })
```

```python
dffiltered.head()
```

```python
# Reorder columns
```

```python
dfcleaned = dffiltered[['race_day','race_name','race_length','race_number_of_finishers','athlete_id','athlete_gender','athlete_age','athlete_performance','athlete_average_speed']]
```

```python
dfcleaned.head()
```

```python
#Visualize data
```

```python
sns.histplot(dfcleaned['race_length'])
```

```python
sns.histplot(dfcleaned, x = 'race_length', hue = 'athlete_gender')
```

```python
sns.displot(dfcleaned[dfcleaned['race_length'] == '50mi']['athlete_average_speed'])
```

```python
sns.lmplot(data = dfcleaned,
           x='athlete_age',
           y='athlete_average_speed', hue='athlete_gender')
```

```python
# Difference in speed for the 50k and 50mi by gender
```

```python
dfcleaned.groupby(['race_length', 'athlete_gender'])['athlete_average_speed'].mean()
```

```python
# What age groups have the best performance in the 50m
```

```python
dfcleaned.query('race_length == "50mi"').groupby('athlete_age')['athlete_average_speed'].agg(['mean','count']).sort_values('mean',ascending = False).query('count>19').head(15)
```

```python
# What age groups have the worst performance in the 50m
```

```python
dfcleaned.query('race_length == "50mi"').groupby('athlete_age')['athlete_average_speed'].agg(['mean','count']).sort_values('mean',ascending = True).query('count>9').head(20)
```

```python
# Seperate data into seasons - find which seasons had better performance

#spring: months 3-5
#summer: months 6-8
#fall: months 9-11
#winter: months 12-2

#split between two decimals
```

```python
dfcleaned['race_month'] = dfcleaned['race_day'].str.split('.').str.get(1).astype(int)
```

```python
dfcleaned['race_season'] = dfcleaned['race_month'].apply(lambda x: 'Winter' if x > 11 else 'Fall' 
                                                         if x > 8 else 'Summer' 
                                                         if x > 5 else 'Spring' 
                                                         if x > 2 else 'Winter')
```

```python
dfcleaned.head(25)
```

```python
dfcleaned.groupby('race_season')['athlete_average_speed'].agg(['mean', 'count']).sort_values('mean', ascending=False)
```

```python
# Show only for 50mi race
```

```python
dfcleaned.query('race_length == "50mi"').groupby('race_season')['athlete_average_speed'].agg(['mean', 'count']).sort_values('mean', ascending=False)
```
